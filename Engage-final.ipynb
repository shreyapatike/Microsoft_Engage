To handle, I grouped similar features together. There are five categories of categorical traits. 'Basic Features' , 'Meters' , 'Advanced Features' , "Null values representing the feature's absence " , "categorical features with a large number of categories".\
'Numericals' (already in numericals) and 'string numericals'  (Feature is numerical but they are represented in strings) are the two types of numerical features.

```python
# Categorical Features #
basic_features = ['Make','Model','Drivetrain','Cylinder_Configuration',
              'Emission_Norm','Engine_Location','Fuel_System','Fuel_Type','Body_Type',
              'Gears','Front_Brakes','Rear_Brakes','Power_Steering','Power_Windows',
              'Keyless_Entry','Seats_Material','Type','12v_Power_Outlet','Audiosystem',
               'Basic_Warranty']

meters = ['Odometer','Speedometer','Tachometer','Tripmeter']

advanced_features =['Cup_Holders','Door_Pockets','Fuel-lid_Opener',
              'Fuel_Gauge','Handbrake','Instrument_Console','Sun_Visor','Third_Row_AC_Vents',
              'Ventilation_System','Adjustable_Headrests','Adjustable_Steering_Column',
              'Parking_Assistance','Infotainment_Screen','Multifunction_Steering_Wheel',
              'Seat_Height_Adjustment','Rear_Center_Armrest','Boot-lid_Opener','Clock']

nulls_are_no = ['Cigarette_Lighter','EBA_(Electronic_Brake_Assist)',
               'Navigation_System','Second_Row_AC_Vents',
               'Turbocharger']

cat_more = ['Front_Suspension','Rear_Suspension','Front_Tyre_&_Rim',
            'Rear_Tyre_&_Rim','Power','Torque','Wheelbase','Wheels_Size',
           'Airbags']

# Numerical Features #
numerical = ['Cylinders','Valves_Per_Cylinder','Doors','Seating_Capacity','Number_of_Airbags']

str_num = ['Ex-Showroom_Price','Displacement','Fuel_Tank_Capacity','Height','Length','Width',
           'ARAI_Certified_Mileage','Kerb_Weight','Ground_Clearance',
          'City_Mileage','Boot_Space','Minimum_Turning_Radius']
```

```python
new_df = pd.DataFrame()
# Created new dataframe to work with our selective features #
```

```python
# Categorical Features #
for feature in basic_features:
    new_df[feature]=df[feature]
    new_df[feature].fillna(new_df[feature].mode()[0],inplace=True) # Replace the null values with mode #
for feature in meters:
    new_df[feature]=df[feature]
    new_df[feature].fillna(new_df[feature].mode()[0],inplace=True) # Replace the null values with mode #
for feature in advanced_features:
    new_df[feature]=df[feature]
    new_df[feature].fillna(new_df[feature].mode()[0],inplace=True) # Replace null values with mode #
for feature in nulls_are_no:
    new_df[feature]=df[feature]
    new_df[feature]=new_df[feature].fillna(value='No') # Replace null values with 'No' #
for feature in cat_more:
    new_df[feature]=df[feature]
    new_df[feature].fillna(new_df[feature].mode()[0],inplace=True) # Replace null values with mode #

 # Numerial Variables #
for feature in numerical:
    new_df[feature]=df[feature]
    new_df[feature].fillna(new_df[feature].mean(), inplace=True) # Replace null values with mean #
for num_feature in str_num:
    new_df[num_feature] = df[num_feature]
    new_df[num_feature].fillna(new_df[num_feature].mode()[0], inplace=True) # Replace null values with mode #
    new_df[num_feature] = new_df[num_feature].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int) # Convert the string into numerical variable #
```

```python
new_df.head()
```

```python
# Encoding of categorical features #
for feature in basic_features:
    new_df[feature]=new_df[feature].astype('category').cat.codes
for feature in meters:
    new_df[feature]=new_df[feature].astype('category').cat.codes
for feature in advanced_features:
    new_df[feature]=new_df[feature].astype('category').cat.codes
for feature in nulls_are_no:
    new_df[feature]=new_df[feature].astype('category').cat.codes
for feature in cat_more:
    new_df[feature]=new_df[feature].astype('category').cat.codes
```

## To predict price

To anticipate the car's pricing, I chose the four most crucial features. \
['Model' , 'Make' , 'Fuel_Type' , 'ARAI_Certified_Mileage']

```python
features_ = ['Model','Make','Fuel_Type','ARAI_Certified_Mileage']
target = ['Ex-Showroom_Price'] # Value to be predicted as a target #

X = df[features_]
# Convert the price from a string to a number. #
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
Y = df['Ex-Showroom_Price']
```

```python
# Used one hot encoder to encode the categorical features #
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoder.fit(X)
```

```python
from sklearn.compose import make_column_transformer
col_shift = make_column_transformer((OneHotEncoder(categories=encoder.categories_),features_),
                                   remainder='passthrough')
```

```python
# Trainig the model using random forest regressor #
Rf =  RandomForestClassifier(random_state=42, n_jobs=-1, max_depth=5,
                                       n_estimators=100, oob_score=True)
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(col_shift,Rf)
pipe.fit(X,Y)
```

```python
# Predict the car's pricing based on user-interest attributes. #
Data = ['Redi-Go','Datsun','Petrol','21.38 km/litre']
pipe.predict(pd.DataFrame(columns=features_,
                          data=np.array(Data).reshape(1,4)))


```
